```python
import pandas as pd
import numpy as np 

nome_do_arquivo = 'dp_env/diabetes.csv'
```

```python
df = pd.read_csv(nome_do_arquivo)
print("Arquivo 'diabetes.csv' carregado com sucesso!")

# Definição das faixas de idade
bins = [0, 20, 40, 60, 80]
labels = ['0-20', '20-40', '40-60', '60-80']
```

```python
# Criar a coluna de faixa etária
df['Faixa_Etaria'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Contagem REAL por faixa
contagem_real = df['Faixa_Etaria'].value_counts().sort_index()

print("\n--- HISTOGRAMA REAL DAS IDADES ---")
print(contagem_real)
```

```python
# ============================
# PRIVACIDADE DIFERENCIAL
# ============================

epsilon = 1.5
sensibilidade = 1   # CORRETO PARA CONTAGEM
b = sensibilidade / epsilon

ruido = np.random.laplace(0, b, size=len(contagem_real))
contagem_dp = contagem_real.values + ruido

# Evitar valores negativos
contagem_dp = np.maximum(contagem_dp, 0)

contagem_dp = pd.Series(contagem_dp, index=contagem_real.index)

print("\n--- HISTOGRAMA COM PRIVACIDADE DIFERENCIAL ---")
print(contagem_dp.round(0))
```
